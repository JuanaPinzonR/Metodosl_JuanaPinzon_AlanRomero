**2.** Una formula alternativa del método de Muller está dada por:
$$
f(x)≅a(x-x_{2})^2+b(x-x_{2})+c
$$
Demuestre que los coeficientes están dados por:
$$
a = \frac{(f[x_{1}; x_{2}]-f[x_{0}; x_{1}])}{(h_{2}-h_{1})}
\\
b = f[x_{1}; x_{2}] + ah_{2}
\\
c = f(x_{2})
$$
donde $h_{1}=x_{1}-x_{0}: h_{2}=x_{2}-x_{1}$. El cero se encuentra en una vecindad de $x_{2}$, en este caso, la formula de Bhaskara es:
$$
x_{3}=x_{2} + \frac{-2c}{b ± \sqrt{b^2 - 4ac}}
$$

Si $b<0$ elegir el signo negativo, de otro modo ($b ≥ 0$), elegir el signo positivo. El criterio de parada es:
$$
\epsilon = |\frac{x_{3}-x_{2}}{x_{3}}| < 1*10^{-10}
$$

Para encontrar estos coeficientes, podemos utilizar las siguientes relaciones:
\
Como la parábola pasa por el punto $(x0, f(x0))$, tenemos que $f(x_{0}) = a(x_{0}-x_{2})^2 + b(x_{0}-x_{2})+c$
\
Como la parábola pasa por el punto $(x1, f(x1))$, tenemos que $f(x_{1}) = a(x_{1}-x_{2})^2 + b(x_{1}-x_{2})+c$
\
Como la parábola pasa por el punto $(x2, f(x2))$, tenemos que $f(x_{2}) = a(x_{2}-x_{2})^2 + b(x_{2}-x_{2})+c=a(0)^2 + b(0)+c=c$


Luego, se pueden restar las 2 priemeras ecuaciones
$$
f(x_{0})-f(x_{1}) = a(x_{0}-x_{2})^2 + b(x_{0}-x_{2})+c - a(x_{1}-x_{2})^2 - b(x_{1}-x_{2})-c\\
f(x_{0})-f(x_{1}) = a[(x_{0}-x_{2})^2-(x_{1}-x_{2})^2] + b[(x_{0}-x_{2})-(x_{1}-x_{2})]\\
f(x_{0})-f(x_{1}) = a[(x_{0}-x_{2}+x_{1}-x_{2})(x_{0}-x_{2}-x_{1}+x_{2})] + b[x_{0}-x_{2}-x_{1}+x_{2}]\\
$$
\
Con esto, se puede reemplazar $h_{1}=x_{1}-x_{0}: h_{2}=x_{2}-x_{1}$
$$
f(x_{0})-f(x_{1}) = a[(h_{1}+h_{2})(h_{1}-h_{2})] + b[x_{0}-x_{1}]\\
f(x_{0})-f(x_{1}) = a[h_{1}^2-h_{2}^2] + b[h_{1}-h_{2}]\\
\frac{f(x_{0})-f(x_{1})}{h_{1}-h_{2}} = a[h_{1}+h_{2}] + b
$$

Si definimos $f[x_{0}; x_{1}] = \frac{f(x_{0}) - f(x_{1})}{h_{1}}$ y $f[x_{1}; x_{2}] = \frac{f(x_{1}) - f(x_{2})}{h_{2}}$, podemos reescribir esta ecuación como:
$$
\frac{f[x_{1}; x_{1}]-f[x_{1}; x_{2}]}{h_{1}-h_{2}} = a[h_{1}+h_{2}] + b
$$
Por un lado, se puede despejar a como
$$
a=\frac{f[x_{1}; x_{1}]-f[x_{1}; x_{2}]}{h_{1}-h_{2}}
$$
Por otro lado, se puede despejar b como:
$$
b=\frac{f[x_{1}; x_{1}]-f[x_{1}; x_{2}]}{h_{1}-h_{2}}-a(h_{1}+h_{2})\\
b=\frac{f[x_{1}; x_{1}]}{h_{1}-h_{2}}-\frac{f[x_{1}; x_{2}]}{h_{1}-h_{2}}-ah_{1}-ah_{2}
$$

Adicionalmente, ya se demostró que $c=f(x_{2})$
