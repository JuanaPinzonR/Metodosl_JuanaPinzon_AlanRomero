## 3.10 Ejercicios: Raíces de polinomios
5. (**SymPy**) Calcular todas las raíces reales de los primeros 20 polinomios de Laguerre. La fórmula de rodrigues es:
$$
L_{n}(x)= \frac{e^x}{n!} \frac{d^n}{dx^n} (e^{-x}x^n)
$$
El intervalo donde existen las raíces es: $0≤x≤\infty$.

```python
# Definir la función de Laguerre
L = Function('L')

# Calcular las raíces reales de los primeros 20 polinomios de Laguerre
for i in range(20):
    n = i
    Ln = L(n, x)
    Lagrangian = Eq(Ln, (exp(x)/factorial(n)) * diff(exp(-x) * x**n, x, n))
    Lagrangian_derivative = diff(Lagrangian.rhs, x)
    Lagrangian_second_derivative = diff(Lagrangian_derivative, x)
    
    # Encontrar las raíces reales
    roots = solve(Lagrangian_second_derivative, x, domain=S.Reals)
    
    # Imprimir las raíces reales
    print(f"Raíces reales de L{n}(x): {roots}")
```

La integral
$$
I=\int_a^b f(x) dx \cong \int_a^b p_{1}(x) dx

Se calcula como sigue:
$$
I=\int_a^b [\frac{x-b}{a-b}f(a) + \frac{x-a}{b-a}f(b)] dx

$$

I=\frac{f(a)}{a-b}(\frac{x^2-2bx}{2}) - \frac{f(b)}{a-b}(\frac{x^2-2ax}{2})

I=\frac{f(a)x^2-2f(a)bx}{2(a-b)} - \frac{f(b)x^2-2f(b)ax}{2(a-b)}

$$

6. Para encontrar la inductancia exacta de una bobina toroidal es necesario calcular la integral:
$$
\int_{-a}^a \frac{\sqrt{a^2-x^2}}{R+x} dx = \pi(R-\sqrt{R^2-a^2})
$$

donde $R = 0.5 cm$ es el radio al centro del toriode y $a = 0.01 cm$ es el radio de la sección transversal del toriode. Estimar la integral con el método del trapecio y la regla de Simpson $1/3$, con un error menor al $0.5%$. Realizar la integral es un verdadero reto analítico.

En primer lugar, se tiene que
$$
f(x) = \frac{\sqrt{a^2-x^2}}{R+x} = \frac{\sqrt{0.01^2-x^2}}{0.5+x}
$$

```python
R=0.5
a=0.01
l = -0.01
m = 0.01
n = 10000000
exact_integral = np.pi * (R - np.sqrt(R**2 - a**2))
```

```python
def f(x):
    return np.sqrt(a**2 - (x)**2) / (R + x)

h = (m - l) / n
```

```python
def trapezoidal_rule(f, l, m, n):
    x = np.linspace(l, m, n + 1)
    y = f(x)
    integral = h * (np.sum(y[0:-1:2]) + 2 * np.sum(y[1::2])) / 3
    return integral

trapecio = trapezoidal_rule(f, l, m, n)

error_relattrap = np.abs(trapecio - exact_integral) / np.abs(exact_integral)

print("Error relativo:", error_relattrap)
print("Integral aproximada:", trapecio)
print("Integral exacta:", exact_integral)
```

### Regla de Simpson 1/3

```python
def simpson_rule(f, l, m, n):
    x = np.linspace(l, m, n + 1)
    y = f(x)
    return h / 3 * (y[0] + y[-1] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2]))

simpson = simpson_rule(f, l, m, n)

error_relatsimp = np.abs(simpson - exact_integral) / np.abs(exact_integral)

print("Error relativo:", error_relatsimp)
print("Integral aproximada:", simpson)
print("Integral exacta:", exact_integral)
```
